## Step 3: Key Concepts

### Backpropagation
The core algorithm for training neural networks. Computes gradients via the chain rule:

$$ \frac{\partial L}{\partial w_{ij}} = \frac{\partial L}{\partial a_j} \cdot \frac{\partial a_j}{\partial w_{ij}} $$

### Activation Functions
- **ReLU** (Rectified Linear Unit): $f(x) = \max(0, x)$ — Default choice, avoids vanishing gradients
- **Sigmoid**: $f(x) = \frac{1}{1+e^{-x}}$ — Used for binary output; vanishing gradient issues in deep nets
- **Softmax**: Used in output layer for multi-class classification

### Loss Functions
- **Cross-Entropy Loss** for classification: $L = -\sum y \log(\hat{y})$
- **MSE** for regression

## Step 4: Visualizing Training Dynamics

A simulated training curve to illustrate convergence behaviour:

````python
epochs = np.arange(1, 21)
train_loss = 2.5 * np.exp(-epochs/5) + 0.05 + np.random.normal(0, 0.02, len(epochs))
val_loss = 2.5 * np.exp(-epochs/6) + 0.10 + np.random.normal(0, 0.03, len(epochs))
train_acc = 1 - np.exp(-epochs/4) * 0.7 + np.random.normal(0, 0.005, len(epochs))
val_acc = 1 - np.exp(-epochs/5) * 0.7 + np.random.normal(0, 0.01, len(epochs))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(epochs, train_loss, 'b-', label='Train', lw=2)
axes[0].plot(epochs, val_loss, 'r-', label='Val', lw=2)
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Loss')
axes[0].set_title('Training Loss', fontweight='bold'); axes[0].legend()

axes[1].plot(epochs, train_acc, 'b-', label='Train', lw=2)
axes[1].plot(epochs, val_acc, 'r-', label='Val', lw=2)
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training Accuracy', fontweight='bold'); axes[1].legend()

plt.tight_layout()
plt.savefig('results/05_training_curves.png', dpi=150, bbox_inches='tight')
plt.show()
````

## Step 8: Reflection

**Why MLP works on MNIST:** Simple structure, low resolution, distinct classes. For more complex images (CIFAR-10, ImageNet), CNN architectures dominate.

**Key takeaways:**
- **ReLU > Sigmoid** in hidden layers (avoids vanishing gradients)
- **Dropout** is crucial for generalization
- **Adam optimizer** with default lr=1e-3 is a strong starting point
- **Batch normalization** can speed up training in deeper networks
